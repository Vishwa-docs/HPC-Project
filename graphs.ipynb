```python
import matplotlib.pyplot as plt
```

## 224x224 - Dependent Filters

```python
x = [1,10,50,100,500,1000,5000,10000,40000,50000]
serial = [0.0045393, 0.010009, 0.035229, 0.0659041, 0.284792,
         0.506872, 2.72359, 5.81001, 23.467,29.0235]
parallel = [0.115256, 0.106761, 0.12881, 0.118222, 0.205095, 0.310721,
            1.1649, 2.42921, 13.9305, 17.7523]

efficiency_d = []
speedup_d = []
for i,j in zip(serial,parallel):
    speedup = (i/j)
    value = ((i/j)/10)*100
    speedup_d.append(speedup)
    efficiency_d.append(round(value, 2))
```

```python
plt.plot(x, serial, label ='serial')
plt.plot(x, parallel, label ='parallel')

plt.title('Dependent Filter')
plt.xlabel("Batchsize")
plt.ylabel("Runtime in secs")
plt.legend()
plt.show()
```
